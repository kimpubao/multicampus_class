```python
def get_category3(age, sex):
    if age < 20:
        return "미성년자"
    elif age < 30:
        return "청년" + ("남성" if sex == "male" else "여성")
    elif age < 50:
        return "중년" + ("남성" if sex == "male" else "여성")
    elif age < 70:
        return "장년" + ("남성" if sex == "male" else "여성")
    else:
        return "노년" + ("남성" if sex == "male" else "여성")

titanic['category3'] = titanic.apply(lambda row: get_category3(row['age'], row['sex']), axis=1)
titanic[['age', 'sex', 'category3']].head()
```

```python
# 1. df1과 df2에 대해 merge 연습, on, left_on, right_on, how
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
df1
df2
```

```python
# 1-1 기본 inner join(기본값) - 공통 키에 대해 병합
pd.merge(df1, df2, on="key")

# 1-2 left join - 왼쪽 DataFrame의 모든 키를 유지
pd.merge(df1, df2, on="key",how="left")

# 1-3 right join - 오른쪽 DataFrame의 모든 키를 유지
pd.merge(df1, df2, on = "key", how = "right")

# 1-4 outer join - 양쪽 DataFrame의 모든 키를 유지
pd.merge(df1,df2, on = "key", how = "outer")

# 1-5 left_on과 right_on을 사용한 병합 - 키가 다른 열 이름을 가진 경우
df3 = pd.DataFrame({
    "key1" : ["a","b","d"],
    "data2" : pd.Series(range(3),dtype="Int64")
})
pd.merge(df1,df3,left_on="key",right_on="key1",how="inner")
```

```python
# 2. left과 right에 대해 merge 연습, on, left_on, right_on, how
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
# 2-1 기본 outer join - 양쪽 DataFrame의 모든 키를 유지
pd.merge(left,right, on=["key1","key2"], how="outer")

# 2-2 inner join - 공통 키에 대해 병합
pd.merge(left,right, on=["key1","key2"],how="inner")

# 2-3 left join - 왼쪽 DataFrame의 모든 키를 유지
pd.merge(left,right,on=['key1','key2'], how = "left")

# 2-4 right join - 오른쪽 DataFrame의 모든 키를 유지
pd.merge(left,right,on=['key1','key2'],how='right')

# 2-5. left_on과 right_on을 사용한 병합 - 키가 다른 열 이름을 가진 경우
left_diff = pd.DataFrame({
    "keyX": ["foo", "foo", "bar"],
    "keyY": ["one", "two", "one"],
    "lval": pd.Series([1, 2, 3], dtype='Int64')
})

right_diff = pd.DataFrame({
    "keyA": ["foo", "foo", "bar", "bar"],
    "keyB": ["one", "one", "one", "two"],
    "rval": pd.Series([4, 5, 6, 7], dtype='Int64')
})

pd.merge(left_diff, right_diff, left_on=["keyX", "keyY"], right_on=["keyA", "keyB"], how="inner")
```

```python
# 3. merge 열 이름과 행 인덱스를 기준으로 병합 연습
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
left1
right1
```

```python
# 3-1 열 이름을 기준으로 병합
pd.merge(left1,right1, left_on='key',right_index=True,how='left')

# 3-2 행 인덱스를 기준으로 병합(인덱스를 리셋하여 열로 사용)
left1_index = left1.set_index("key")
left1_index.join(right1, how="left")

# 3-3 인덱스를 기준으로 직접 병합
left1_with = left1.set_index('key')
left1_with.join(right1,how='left')
```

```python
# 4. merge연습, index 기준
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=["a", "c", "e"],
                     columns=["Ohio", "Nevada"]).astype("Int64")
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=["b", "c", "d", "e"],
                      columns=["Missouri", "Alabama"]).astype("Int64")
left2
right2
```

```python
# 4-1 기본 inner join - 공통 인덱스에 대한 병합
left2.join(right2,how='inner')

# 4-2 left join - 왼쪽 DataFrame의 모든 인덱스를 유지
left2.join(right2,how='left')

# 4-3 right join - 오른쪽 DataFrame의 모든 인덱스를 유지
left2.join(right2, how='right')

# 4-4 outer join - 양쪽 DataFrame의 모든 인덱스를 유지
left2.join(right2, how='outer')
```

```python
# 5-1. concat연습, axis,join 속성 값 변경 설정
s1 = pd.Series([0, 1], index=["a", "b"], dtype="Int64")
s2 = pd.Series([2, 3, 4], index=["c", "d", "e"], dtype="Int64")
s3 = pd.Series([5, 6], index=["f", "g"], dtype="Int64")
```

```python
# 5-1-1 기본 concat - 수직으로 병합(axis=0, 기본값)
pd.concat([s1,s2,s3])
pd.concat([s1,s2,s3], axis=0)

# 5-1-2 수평으로 병합(axis=1)
pd.concat([s1,s2,s3], axis=1)

# 5-1-3 수평으로 병합 / Inner join(공통 인덱스만 병합)
pd.concat([s1, s2, s3], axis=1, join='inner')

# 5-1-4 수평으로 병합, outer join(모든 인덱스를 포합)
pd.concat([s1, s2, s3], axis=1, join='outer')

# 5-1-5 keys 사용하여 수직으로 병합 (multiindex)
pd.concat([s1, s2, s3], keys=['s1', 's2', 's3'])

# 5-1-6 keys 사용하여 수평으로 병합 (multiindex,axis=1)
pd.concat([s1, s2, s3], axis=1, keys=['s1', 's2', 's3'])
```

```python
# 5-2. concat연습, axis,join 속성 값 변경 설정
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=["a", "b", "c"],
                   columns=["one", "two"])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=["a", "c"],
                   columns=["three", "four"])
df1
df2
```

```python
# 5-2-1. 기본 concat - 수직으로 병합 (axis=0, 기본값)
pd.concat([df1, df2])

# 5-2-2. 수평으로 병합 (axis=1)
pd.concat([df1, df2], axis=1)

# 5-2-3. 수평으로 병합, inner join (공통 인덱스만 병합)
pd.concat([df1, df2], axis=1, join='inner')

# 5-2-4. 수평으로 병합, outer join (모든 인덱스를 포함)
pd.concat([df1, df2], axis=1, join='outer')

# 5-2-5. keys 사용하여 수직으로 병합 (MultiIndex)
pd.concat([df1, df2], keys=['df1', 'df2'])

# 5-2-6. keys 사용하여 수평으로 병합 (MultiIndex, axis=1)
pd.concat([df1, df2], axis=1, keys=['df1', 'df2'])
```

```python
# 6-1 stack - 열을 행으로 전환
data.stack()

# 6-2 unstack - 행을 다시 열로 전환
data.unstack()


# 6-3 특정 레벨을 기준으로 stack
data.unstack(level="state")

# 6-4 
data.stack(level="number")
```

```python
# 7. 타이타닉
'''
7-1 survived열 값으로 전체 생존율 출력
7-2 pclass열 값이 1인 자료의 개수를 출력
7-3 age가 10보다 작은 승객에 대한 생존율
7-4 embarked에서 각 항구 별 승선인원 출력

'''
titanic = sns.load_dataset("titanic")
titanic

# 7-1. survived 열 값으로 전체 생존율 출력
survived = titanic['survived'].mean()*100
print('\n전체 생존율 :', '{:.1f}'.format(survived),'%')

# 7-2 pclass열 값이 1인 자료의 개수를 출력
pclass_count = titanic[titanic['pclass'] == 1].shape[0]
print('\npclass열 값이 1인 자료의 개수 :', pclass_count)

# 7-3 age가 10보다 작은 승객에 대한 생존율
ten_under = titanic[titanic['age'] < 10]
ten_under_mean = ten_under['survived'].mean()*100
print('\n10세미만 승객의 생존율','{:.1f}'.format(ten_under_mean),'%')

# 7-4 embarked에서 각 항구 별 승선인원 출력
embarked_count = titanic['embarked'].value_counts()
print('\n각 항구 별 승선인원')
print(embarked_count)
```
