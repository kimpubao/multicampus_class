```python
import FinanceDataReader as fdr
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = fdr.DataReader('005930','2015-01-01','2024-07-03')
df
```

```python
df['Close'].plot()
```

```python
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    return numerator / (denominator + 1e-7)
    # 0으로 나누기 에러 방지하기 위해 매우 작은 값을 더해서 나눔
```

```python
df
```

```python
dfx = df[['Open','High','Low','Volume', 'Close']]
dfx = MinMaxScaler(dfx)
```

```python
dfx
```

```python
dfy = dfx[['Close']]
dfx = dfx[['Open','High','Low','Volume']]
dfx
```

```python
dfx.describe()
```

```python
# 지난 30일간의 데이터로 오늘의 주가를 예측
# 5월 25일 ~ 7월 3일 까지 30일간의 삼성전자 주가 데이터로 7월 4일 종가 예측
# 5월 26일 ~ 7월 4일 까지 30일간의 삼성전자 주가 데이터로 7월 5일 종가 예측
```

```python
X = dfx.values.tolist()
y = dfy.values.tolist()
```

```python
X[:2]
```

```python
y[:2]
```

```python
window_size = 30

data_X = []
data_y = []
for i in range(len(y) - window_size): #2338-30=2308=ramge(2308) = i:2307
    _X = X[i : i + window_size] # 30일간 주가 데이터(1월1일~1월30)
    _y = y[i + window_size]   #31일째의 종가 데이터(다음날, 1월31일)
    data_X.append(_X) # 30일간 주가 데이터 단위로 data_X에 저장
    data_y.append(_y) # 다음 날의 종가 데이터가 data_y에 저장
print(_X, "->", _y)
```

```python
train_size = int(len(data_y) * 0.7)
train_X = np.array(data_X[0 : train_size])
train_y = np.array(data_y[0 : train_size])

test_size = len(data_y) - train_size
test_X = np.array(data_X[train_size : len(data_X)])
test_y = np.array(data_y[train_size : len(data_y)])

print('훈련 데이터의 크기 :', train_X.shape, train_y.shape)
print('테스트 데이터의 크기 :', test_X.shape, test_y.shape)
```

```python
# 2338건 주가 데이터 -> 2308개의 훈련데이터 -> 7:3으로 비율로 나눔
# 훈련 데이터의 크기 : x=(1615, 30, 4) y=(1615, 1)
# 테스트 데이터의 크기 : (693, 30, 4) (693, 1)

# (693, 30, 4)를 x데이터로 (1615, 1)를 y데이토로 하여 훈련진행
# => model => model로 (693, 30, 4)데이터를 예측했을 때 (693,1)과 비교하여
# 모델에ㅐ 대한 성능 평가
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
model = Sequential()
```

```python
model.add(LSTM(units=20, return_sequences=False, input_shape=(30,4)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(train_X, train_y, epochs=70, batch_size=40)
pred_y = model.predict(test_X)
```

```python
pred_y = model.predict(test_X)

plt.figure()
plt.plot(test_y, color='red', label='real SEC stock price')
plt.plot(pred_y, color='blue', label='predicted SEC stock price')
plt.title('SEC stock price prediction')
plt.xlabel('time')
plt.ylabel('stock price')
plt.legend()
plt.show()

print("내일 예상되는 삼성전자 주가 :", df.Close[-1] * pred_y[-1] / dfy.Close[-1], 'KRW')
```
