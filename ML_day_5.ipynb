k-평균 알고리즘
- 주어진 데이터를 k개의 클러스터로 묶는 비지도 학습 알고리즘
- 데이터의 특징만으로 비슷한 데이터끼리 모아 군집화 된 클래스로 분류

k-평군 알고리즘 계산 순서


k값/중심설정 -> 데이터의 클러스트 결정 
-> 클러스터 내 중심 이동 -> 중심이 변경되지 않을 때까지 2, 3번 과정 반복

k값 적정값 구하는법
- 시각화해 보고 적정한 값을 넣어준다

k값 / 최초 중심 설정
- k는 몇개의 클래스로 분류할 것인지 나타내는 변수(k=3)
- -> k값에 따라 모델의 성능이 좌우

- 최초 중심 설정 방법
1. 무작위로 설정(잘 사용하지 않음)
2. 사용자가 직접 설정(고차원일때 잘 구현하기 어려워서 잘 용하지 않음)
3. k-means++

클러스터 내 중심 이동
- 데이터의 순회가 완료되면 각 클러스터의 중심값을 실제 클러스터 내부의 데이터들의 중아값으로 변경


k-평균 알고리즘 장단점

장점
- 감단한 알고리즘으로 구현이 용이
- 계산시간이 비교적 짧음(속도가 빠름)

단점
- 초기 클러스터 개수와 중심에 따라 결과 크게 달라질 수 있음
- 결과 해석이 쉽지 않음

계층적 군집화
- 계층적 트리 모형을 이용해 개별 개체들을 순차적, 계층적으로 유사한 그룹과 동합하여 군집화를 수행하는 알고리즘
- 군집 간의 거리를 기반으로 군집화를 수행

군집화 방식
1. 단일기준 결합 방식
- 그룹 사이에서 가까운 거리에 있는 것들 계산하는 방식
2. 완전 기준 결합 방식
- 그룹 사이에서 멀리 떨어져 있는 거리에 있는 것들 계산하는 방식
3. 평균 기준 결합 방식
- 그룹 사이에서 평균 거리에 있는 것들 계산하는 방식

계층적 군집화 관련 용어
1. 덴드로그램
- 표준들이 군을 형성하는 과정을 나타내는 나무 형식의 그림
- 군집 수는 사용자가 직접 정할 수 있음

2. 군집의 결정
- 적절한 수준에서 트리를 잘라서 클러스터 수를 조절
- 군집 수를 사전에 지정하지 않아도 됨

클러스터링 대상인 데이터 포인트(data points) 
- 데이터 포인트는 여러 속성(attributes)으로 구성되어 있음 
- 데이터 포인트들 간에는 유사 척도(similarity measure)가 부여됨 (가깝고 먼 것들을 구분할 수 있는 거리 척도가 존재함) 


클러스터링이란?
- 클러스터 내의 데이터 포인트들은 서로 많이 유사한  반면
- 클러스터 간의 데이터 포인트들은 서로 적게 유사하도록
- 데이터 포인트들을 클러스터들로 그룹핑하는 작업

유사 척도(유사도, similarity measure)
- 속성이 연속(continuous)인 경우 유클리디안 거리(Euclidean distance)를 사용
- 그렇지 않은 경우, 해당 문제에 적합한 척도를 (정의하여) 사용함 

문서 클러스터링 (document clustering) 

- 목적: 문서에 나타나는 주요 용어(important terms)를 기반으로, 서로 서로 유 사한 문서들의 그룹을 찾아라. 
  
- 접근법 
1. 문서들에 자주 나타나는 용어들을 식별한다. 

2. 용어들의 출현 빈도를 기반으로 유사 척도를 정의한다. (즉, 어떤 용어가 공통으로 자주 나타나면 유사한지를 유사 척도로 정의한다.)  
3.  정의한 유사 척도를 사용하여 문서들을 클러스터링 한다.  
- 정보 검색(information retrieval) 활용: 새로운 문서나 새로운 검색어를 이미 클러스터된 문서들을 관련 짓는데 활용할 수 있다.

선형회귀 알고리즘

회귀
- 예상하지 못한 현상이 일어나더라도 그 현상은 결국 어떤 특정 예상 값으로 돌아간다는 의미
- 두 변수의 관계사 어떤 일반화된 선형관계의 평균으로 돌아간다(회귀)의미

선형회귀
- 관찰된 데이터들을 기반으로 하나의 함수를 구해거 관찰되지 않은 데이터 값을 예측하는 것

- 과거의 데이터 -> 일정한 규칙 탐색 -> 미래예측

일차 함수 y = αx + β

- 일차 함수 y = αx + β의 그래프는

- 일차 함수 y = αx의 그래프를 y축의 방향으로 β만큼 평행이동한 직선

선형회귀 모델
- 회귀 계수를 선형적으로 결합할 수 있는 모델을 의미

독립변수(x)
- 어떤 것의 원이이 되는 변수(영향을 주는 변수)

종속변수(y)
- 독립변수의 변화에 따른 결과를 나타내는 변수(영향을 받는 변수)

회귀 계수
- 독립변수의 변화가 종속변수에 미치는 영향력 크기를 나타내는 변수
- y = β0 + β1X

단일 회귀분석
- x, y하나인 것
- 광고비 = x / 매출 =y

다중 회귀분석
- x는 여러가지 Y는 하나
- 광고비 x1, 교육비 x2, 유통비 x3 /  매출 y
- x1+x2+x3 = y

회귀 함수 구하기
- 선형 회귀 분석에서 가장 중요한 단계
- 주어진 데이터로부터 가장 알맞은(오차의 합이 최가 되는) 함수를 구하는 것


- y=βx라는 한가지의 회귀계수만을 사용한 함수로 표현해야 한다고 가정할 경우
- -> 가장적당한 β구해야한다

- 오차 : 데이터 포인트에서 함수의 선까지의 거리(차이) -> 합계
- 실무에서는 오차값을 제곱하여 더 많이 사용

평균 제곱 오차(MSE,mean squared erroer)
- 각 데이터포인트로 부터 함수까지의 거리를 제곱한 값을 평균 낸 값

목적 함수
- 평균 재곱오차를 최소화하는 함수
- 사각형들을 최소로 하는 회귀함수를 찾는 것이 목적
- 회귀의 목적 함수는 평균 제곱 오차를 최소화하는 것

경사 하강법
- 함수의 기울기(경사)를 구하고 경사의 절댓값이 낮은 쪽으로 계속 이동시켜 극 값에 이를 때까지 반복하는 방법
- 파라미터를 임의로 설정 후, 조금씩 변화시키면서 손실을 최소화하는 방향으로 최적의 파라미터를 찾아 나감

목적 함수 구하기
- y = βx라는 회귀함수의 B에 따른 평균 제곱 오차 그래프(β -> w)
- 평균 제곱 오차를 최소를 하는 w는 변곡점 부분이며, 기울기(미분값)이 0인 지점
- 경사 하강법을 반복 수행하면서 미분 값이 0인 w를 구함
- 평균 제곱 오차가 최소화된 β를 회귀함수의 회귀계수로 선택

학습률
- 학습률이 너무 낮을 경우 -> 학습속도가 너무 느림
- 학습률이 너무 높을 경우 -> 학습이 제대호 이루어지지 않음

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas_datareader as pdr
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import datasets
from sklearn.cluster import KMeans


# 맥 폰트 깨짐 이슈 해결
from matplotlib import rc
%matplotlib inline
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
labels = pd.DataFrame(iris.target)
```

```python
labels
```

```python
labels.columns=['labels']
```

```python
data = pd.DataFrame(iris.data)
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
```

```python
data
```

```python
data = pd.concat([data,labels],axis=1)
data
```

```python
feature = data[ ['Sepal length','Sepal width']]
feature.head()
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(feature)
```

```python
feature
```

```python
model.predict(feature)
```

```python
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']
```

```python
r = pd.concat([feature,predict],axis=1)
r
```

```python
model.cluster_centers_
```

```python
centers = pd.DataFrame(model.cluster_centers_,columns=['Sepal length','Sepal width'])
center_x = centers['Sepal length']
center_y = centers['Sepal width']
```

```python
centers
```

```python
plt.scatter(r['Sepal length'],r['Sepal width'],c=r['predict'],alpha=0.5)
plt.scatter(center_x,center_y,s=50,marker='D',c='r')
plt.show()
```
