```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()

# 1.타이타닉호 승객 데이터의 데이터 개수를 각 열마다 구해본다.
print('각 열의 데이터 수\n',titanic.count())
```

```python
# 2.sort_values 메서드를 사용하여 타이타닉호 승객에 대해 성별(sex) 인원수, 나이별(age) 인원수, 선실별(class) 인원수, 사망/생존 인원수를 출력하시오.
# 2-1 성별 인원수
sex_count = titanic['sex'].value_counts().sort_values(ascending=False)
print('\n성별 인원 수\n',sex_count)

# 2-2 나이별 인원수
age_count = titanic['age'].value_counts().sort_values(ascending=False)
print('\n나이별 인원 수\n',age_count)

# 2-3 선실별 인원수
class_count = titanic['pclass'].value_counts().sort_values(ascending=False)
print('\n선실별 인원 수\n',class_count)

# 2-4 사망/생존 인원수
survived_count = titanic['survived'].value_counts().sort_values(ascending=False)
print('\n사망/생존 인원 수\n',survived_count)
```

```python
# 3.타이타닉호 승객의 평균 나이를 구하라.

mean_age = titanic['age'].mean()
print("\n승객의 평균 나이 :", mean_age)
```

```python
# 4.타이타닉호 승객중 여성 승객의 평균 나이를 구하라.

female_mean_age = titanic[titanic['sex'] == 'female']['age'].mean()
print("승객의 평균 나이 :", female_mean_age)
```

```python
# 5.타이타닉호 승객중 1등실 선실의 여성 승객의 평균 나이를 구하라.
female_firstclass_mean_age = titanic[(titanic['sex'] == 'female') & (titanic['pclass'] == 1)]['age'].mean()
print('1등실 여성 승객의 평균 나이 :',female_firstclass_mean_age)
```

```python
# 6.타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만들어라. category1 카테고리는 다음과 같이 정의된다.
# - 20살이 넘으면 성별을 그대로 사용한다.
# - 20살 미만이면 성별에 관계없이 “child”라고 한다.

titanic['category1'] = titanic['age'].apply(lambda x: 'child' if x < 20 else 'male' if x >= 20 and pd.notnull(x) else 'unknown')
titanic[['age', 'sex', 'category1']].head()
```

```python
# 7.타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고쳐라.

mean_age = titanic['age'].mean()
titanic['age'].fillna(mean_age, inplace=True)
mean_age
```

```python
# 8. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category2 열을 만들어라. category2 카테고리는 다음과 같이 정의된다.
# - 성별을 나타내는 문자열 male 또는 female로 시작한다.
# - 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열이 온다.
# *예를 들어 27살 남성은 male27 값이 된다.

titanic['category2'] = titanic['sex'] + titanic['age'].astype(int).astype(str)
titanic[['sex', 'age', 'category2']].head()
```

```python
# 9.타이타닉호 승객을 ‘미성년자’, ‘청년’, ‘중년’, ‘장년’, ‘노년’ 나이 그룹으로 나눈다.
# bins = [1, 20, 30, 50, 70, 100]
# labels = ["미성년자", "청년", "중년", "장년", "노년"]
# 그리고 각 나이 그룹의 승객 비율을 구한다. 비율의 전체 합은 1이 되어야 한다.

bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels, right=False)
age_group_proportion = titanic['age_group'].value_counts(normalize=True)
age_group_proportion
```

```python
# 10. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category3 열을 만들어라. category3 카테고리는 다음과 같이 정의된다.
# - 20살 미만이면 성별에 관계없이 “미성년자”라고 한다.
# - 20살 이상이면 나이에 따라 “청년”, “중년”, “장년”, “노년”을 구분하고 그 뒤에 성별을 나타내는 “남성”, “여성”을 붙인다.
def get_category3(age,sex):
    if age < 20:
        return '미성년자'
    elif age <30:
        return '청년' + ('남성' if sex == 'male' else '여성')
    elif age < 50:
        return '중년' + ('남성' if sex == 'male' else '여성')
    elif age < 70:
        return '장년' + ('남성' if sex == 'male' else '여성')
    else:
        return '노년' + ('남성' if sex == 'male' else '여성')
    
titanic['category3'] = titanic.apply(lambda row: get_category3(row['age'], row['sex']), axis=1)
titanic[['age', 'sex', 'category3']].head()
```
