머신러닝 학습기법 : 지도학습 / 비지도학습 / 강화학습

1. 지도학습 (Supervised Learning)
- 지도학습은 입력 데이터와 그에 대한 레이블 또는 정답을 함께 사용하여 모델을 학습시키는 방법
- 모델은 입력과 출력 간의 관계를 학습하고, 새로운 입력 데이터에 대한 예측을 수행할 수 있음
- 예를 들어, 과거의 주택 가격 데이터와 해당 주택의 평방 피트 면적을 가지고 있는 경우, 지도학습을 사용하여 새로운 주택의 가격을 예측할 수 있음

​

2. 비지도학습 (Unsupervised Learning)
- 비지도학습은 레이블이 없는 입력 데이터만을 사용하여 모델을 학습시키는 방법
- 모델은 데이터의 내부 구조, 패턴, 군집 등을 발견하고 이해하는 데 중점을 둠
- 예를 들어, 고객의 구매 이력 데이터만 가지고 있다면, 비지도학습을 사용하여 비슷한 구매 패턴을 가진 고객들을 군집화하거나, 데이터의 차원을 축소하는 등의 작업을 수행할 수 있음

​

3. 강화학습 (Reinforcement Learning):
- 강화학습은 에이전트가 주어진 환경 속에서 행동을 취하고 이에 따른 보상을 받으며, 시간이 지나면서 최적의 행동을 학습하는 방법
- 에이전트는 시행착오를 통해 보상을 최대화하기 위한 효율적인 행동 방법을 스스로 학습
- 예를 들어, 로봇이 미로에서 목적지까지 이동해야 하는 경우, 강화학습을 사용하여 로봇이 최적의 경로를 학습하고 목적지에 도착할 수 있도록 할 수 있음

1. 과소적합

   모델 학습시, 충분하지 못한 특징만으로 학습되어,

   0) 특정 특징에만 편향되게 학습된것(편향이 높음)

   1) 테스트 데이터뿐만 아니라 학습 데이터에 대해서도 
      정확도가 낮게 나올 경우 과소 적합된ㄴ 모델일 가능성이 높음

   2) 학습에 사용된 특징의 개수를 늘리는 방법을 통해 개선


2. 과대 적함

   학습데이터에 대한 정확도ㅇ는 매우 높지만 테스트 데이터 또는

   0) 학습 데이터 외의 데이터에는 정확도가 낮개 나오는 것(분산이 높음)

   1) 특징이 필요 이상으로 많을 경우 발생

   2) 훈련데이터를 더 많이 모으거나
      학습에 사용된 특징의 개수를 줄이는 방법을 통해 개선

머신러닝 모델의 성능 평가
1. 정확도
   1) 입력된 데이터에 대해 올바르게 예측한 비율
   2) 혼동 행렬 상에서는 대각선을 전체 셀로 나눈 값에 해당

2. 정밀도
   1) 모델의 예측 값이 얼마나 정확하게 예측됐는가를 나타내는 지표
   2) False를 True라고 판단하면 안되는 경우 중요

3. 재현울
   1) 실제 값 중에서 모델이 검출한 실제 값의 비율을 나타내는 지표
   2) True룰 False로 잘못 판단하면 큰일나는 경우 중요
   3) 정밀도와 재현율운 trade-off관계


1. k-폴드 교차 검증
   - 학습데이터의 일정부분을 검증 데이터로 쓰되 k번의 검증 과정을 통해 학습 데이터의 모든 데이터를 반번씩 검증 데이터로 사용하는 방식

Training set / Validation Set / Test Set의 개념 구분

2. 장점
   1) 검증 결과가 일정데이터에 치우치지 않고 
      모든 데이터에 대한 결과이므로 신빙성이 높음

   2) 별도로 검증 데이터를 분리하지 않아도 됨

```python
train_df = pd.read_csv('titanic/train.csv')
test_df = pd.read_csv('titanic/test.csv')
```

```python
# Parch (부모/자녀 수) 값별 개수 계산
parch_counts = train_df['Parch'].value_counts().sort_index()
parch_dict = parch_counts.to_dict()

# Embarked (탑승한 항구) 값별 개수 계산
embarked_counts = train_df['Embarked'].value_counts().sort_index()
embarked_dict = embarked_counts.to_dict()

# Cabin (상위 5개 객실) 값별 개수 계산
cabin_counts = train_df['Cabin'].value_counts().head()
cabin_dict = cabin_counts.to_dict()

print(parch_dict)
print(embarked_dict)
print(cabin_dict)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Parch (부모/자녀 수) 바 차트
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Parch')
plt.title('탑승한 부모/자녀 수')
plt.show()

# Embarked (탑승한 항구) 바 차트
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Embarked')
plt.title('탑승한 항구')
plt.show()

# Cabin (상위 5개 객실) 바 차트
plt.figure(figsize=(10, 6))
cabin_counts.plot(kind='bar')
plt.title('객실 top5')
plt.show()

# 생존 비교 바 차트
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Survived')
plt.title('생존자 수')
plt.show()
```

```python
# 이름에서 호칭 추출하여 Title 열 생성
train_df['Title'] = train_df['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())

# 각 호칭의 개수를 확인
title_counts = train_df['Title'].value_counts()

# 호칭이 1-2개인 것들을 'Other'로 그룹화
rare_titles = title_counts[title_counts <= 2].index
train_df['Title'] = train_df['Title'].replace(rare_titles, 'Other')

# Title 인코딩
title_mapping = {'Mr': 0, 'Miss': 1, 'Mrs': 2, 'Master': 3, 'Dr': 4, 'Rev': 5, 'Other': 6}
train_df['Title'] = train_df['Title'].map(title_mapping)

# 각 호칭의 카운트 계산
train_df['Title'].value_counts()
```
