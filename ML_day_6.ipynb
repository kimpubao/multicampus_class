- 선형회귀 -> 로지스틱 회귀 -> softmax -> N/N -> DNN(딥러닝) -> RNN / CNN

주성분 분석
- 고차원의 데이터를 저차원의 데이터로 차원 축소하는 알고리즘
- 주로 고차원의 데이터를 3차원 이하의 데이터로 바꿔서 시각화하는 데 많이 사용
- 우용한 정보만 살려서 적은 메모리에 저장하거나 데이터의 노이즈를 줄이고 싶을 때도 사용

주성분 분석의 원리
- 데이터의 분산을 최대한 유지하면서 저차원으로 데이터를 변환하는 것
- 분산을 유지하는 이유는 데이터의 고유한 특성을 최대한 유지하기 위함
- 주성분 분석 알고리즘은 수학적인 방법으로 데이터 정보의 유실이 가장 적은 라인을 찾아냄
- 수학적으로 ‘데이터 중첩이 가장 적다’라는 말은 ‘데이터의분산이 가장 크다’라는 말과 동일

로직스틱 회귀
- 묻는 말에 예 / 아니오로만 답하는 것
- 데이터가 어떤 범주에 속할 확률을 0애서 1사이의 값으로 예측하고 그 확률에 따라 가능성이 더 높은 범주에 속하는 것으로 분류하는 이진 분류 모델
- 
로지스틱 회귀 활용 예시)
- 스팸메일 -> 머신러닝 -> 지금 수신한 메일은 스팸일 확률이 0.6이므로 스팸메일로 분류
- 스팸메일 -> 머신러닝 -> 지금 수신한 메일은 스팸일 확률이 0.2이므로 일반메일로 분류
- 

시그모이드 함수 
- 확률은 0~1 사이의 S자형 곡선 모양으로 나타내주는 함수

로지스틱 회귀의 학습
- 현재 가지고 있는 데이터를 통해 에러를 줄이는 방향으로 w와 b의 최적값을 찾아내는 과정

0, I 인코딩
- 스팸 감지: 스팸 (I) 또는 메일(0)
- 페이스북 피드: 보여주기(I) 또는 숨기기(0)
- 신용 카드 사기 거래 감지: 정당(0) 또는 사기(1)

선형 회귀
- 우리는 Y가 0 또는 1이라는 것을 알고 있음
- 가설은 1보다 크거나 0보다 작은 값을 가질 수 있음

크로스 엔트로피

비용함수(Cost function)찾기
- 로지스틱 회귀도 경사 하강법을 사용하여 가중치를 찾아냄
- 비용 함수로 평군 제곱 오차를 활용하지 않음
- ->로컬 미니엄에 빠질 수 있기 때문

- 크로스엔트로피를 비용함수로 최솟값을 찾아 나갈 때 실제 값이 0,1에 상관 없이 한쪽으로만 값이 수렴
- 이러한 단순함은 경사하강법이 더욱 쉽게 최적의 w찾는 데 도움

다중 분류 로지스틱 회귀
- 로지스틱 회귀를 여러 개 붙여서 정규화된 출력을 함으로써 여러 개의 값 분류 가능
- 보통 소프트맥스라고 함


앙상블 기법의 정의

앙상블 기법
- 여러 개의 분류기를 생성하고, 그 예측을 결합하여 더욱 정확한 예측을 도출하는 기법
- 강력한 하나의 모델을 사용 < 약한 모델 여러 개를 조합하여 사용해 강한 모델 생성

배깅(Bagging)
- 샘플을 여러번 추출 -> 모델을 학습 -> 결과물을 집계
- 알고리즘의 안정성 및 정확성 향상을 위해 수행
- 부트스트랩(bootstrap)과 어그리게이팅(aggregating)을 합친 단어
- 대표적인 모델 : 랜덤포레스트(Random Forest)


부트스트랩
- 분류 모델을 여러 개 만들어서 서로 다른 학습 데이터로 학습
- 데이터를 약간 편향되도록 샘플링하는 기법
- 의사결정 트리처럼 과대적합되기 쉬운 모델을 앙상블할 때 사용
- 데이터 샘플링 시 편향을 높임으로써 분산이 높은 모델의 과대적합 위험을 줄이는 효과를 가져옴
- 각 분류 모델을 총 N개의 데이터보다 적은 데이터로 학습하더ㅣ, 중복된 데이터를 허용함으로써 편향이 높은 학습 데이터로 학습
- 데이터 샘플링 크기는 보통 전체 데이터의 60-70%를 사용

어그리게이팅
- 서로 다른 예측 값들을 투표를 통해 가장 높은 예측 값으로 결정
- 여러 분ㄹ휴 모델이 예측한 값들을 조합해서 하나의 결론을 도출하는 과정
- 결론은 투표를 통해 결정

하드 보팅
- 배킹에 포함된 K개의 분류 모델에서 최대 득표를 받은 예측값으로 결론 도출

소프트 보팅
- 각 분류값 별 확률을 합한 값을 점수로 사용해 최대 점수를 가진 분류값을 결론으로 도출

랜덤 포레스트
- 다수의 결정 트리들을 배깅해서 예측을 실행하는 앙상블 기법
- 각 노드에 주어진 데이터를 샘플링해서 일부 데이터를 제외한 채최적의 특징을 찾아 트리를 분기
- 모델의 편향을 증가시켜 과대적합의 위험을 감소
- 다양한 분야에서 비교적 좋은 성능을 보여줌

-  배깅을 이용한 포레스트 구성
	- 트리들이 서로 조금씩 다른 특성을 갖게 되어 일반화 성을 행상할 수 있음

부스팅
- 이전 분류기의 학습 결과를 바탕으로 다음 분류기의 학습 데이터의 샘플 가중치를 조정하여 학습하는 방법
- 동일한 할고리즘의 분류기를 순차적으로 학습해서 여러 개의 분류기를 만든 후, 테스트할 때 가중 투표를 예측값을 결정
- 대표적 모델 : XGBoost와 AdaBoost, GrandientBoost

- 예제
- 1. 인물 사진 안에 있는 인물을 보고, 남자 또는 여자로 분류하는 의셜정 트리블 부그팅
- 2. 테스트 결과, 남자 분류가 미흡할 경우, 남자 학습 데이터를 보강한 후 두번째 의사 결정 트리를 학습
- 3. 두번째 의사 결정 트리의 데스트 결과에 따라 학습 데이터를 보강해서 세번째 의사 결정 트리를 학습


가중 투표 - 하드 보팅
- 분류기의 성능에 따라 가중치를 반영하여 투표 결과를 조장

랜덤 포레스트 주요 파라미터 
- n_estimators 모델에서 사용할 트리 개수 
- max_features 최적의 분할을 위해 고려할 피처의 최대 개수 
- max_depth 트리의 최대 깊이 
- min_samples_leaf 리프 노드가 되기 위한 최소 샘플 데이터 개수 
- min_samples_split 자식 노드를 갖기 위한 최소한의 데이터 개수

```python
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i,init='k-means++' ,max_iter=300,random_state=0)
    kmeans.fit(scaled_feature)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss) 
plt.title('THe Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel("WCSS") #WCSS stands for total within-cluster sum of sqaure
plt.show()
```
